```python
df1=df.copy()
#df1.datetime=df1.datetime.astype('datetime64')
#applymap(df1.datetime)
```

```python
# date time is in object type format, need to convert it into datetime64
```

```python
df.datetime = df.datetime.astype('datetime64')
```

```python
df.datetime.dtype
```

```python
df
```

```python
# adding the new columns
```

```python
df['date'] = df.datetime.apply(lambda x : x.date().strftime('%m/%d/%Y'))  # strftime to change to format dd/mm/yyyy, it is dd-mm-yyyy by default
df['Month'] = df.datetime.apply(lambda x : calendar.month_name[x.month])
df['weekday'] = df.datetime.apply(lambda x : calendar.day_name[x.weekday()])
df['Hour'] = df.datetime.apply(lambda x : x.hour)
```

```python
from sklearn.preprocessing import OneHotEncoder

cols = ['season', 'weather', 'Month', 'weekday']

ohe = OneHotEncoder(sparse=False)

encoded = ohe.fit_transform(df[cols])
encoded = pd.DataFrame(encoded)
encoded.index = df.index
encoded.columns = ohe.get_feature_names(cols)

df.drop(cols,inplace=True,axis=1)
encoded = pd.DataFrame(encoded)
df = pd.concat([df,encoded],axis=1)

df.head()
```
