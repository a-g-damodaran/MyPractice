```python
# Employee Records for HRA & DA
emp = dict()
emp['E_No'] = []
emp['E_Name'] = []
emp['E_BPay'] = []
l1=[]
l2=[]
l3=[]
n = int(input('No of Employee - '))
for i in range(1,n+1):
  l1.append(input('Enter Emp No :'))
  l2.append(input('Enter Emp Name :'))
  l3.append(int(input('Enter Basic Pay :')))
emp['E_No'] = l1
emp['E_Name'] = l2
emp['E_BPay'] = l3
print(emp)
```

```python
import pandas as pd
l4 = []
l5 = []
for i in range(len(emp['E_BPay'])):
  if emp['E_BPay'][i] <=15000:
    l4.append(emp['E_BPay'][i]*.1)
    l5.append(emp['E_BPay'][i]*.05)
  elif emp['E_BPay'][i] > 15000:
    l4.append(emp['E_BPay'][i]*.07)
    l5.append(emp['E_BPay'][i]*.03)
emp['HRA'] = l4
emp['DA'] = l5
print(emp)
```

```python
# Prime Number
n = int(input())

while j <= n:
  j = 0
  cnt = 0
  for i in range(2,n):
    if n%i == 0:
      #print(n,'np')
      break
    else:
      cnt += 1
  if cnt == (n - 2):
    print(n,'p')
    j += 1
  n -= 1
```
